```python
## 3D2
# f(x,y) = sqrt(x^2+y^2)

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mpl_toolkits.mplot3d import axes3d

plt.rcParams['figure.figsize'] = [6, 5]

num = 301
a = 2
x = np.linspace(-3*a, 3*a ,num)
y = np.linspace(-3*a, 3*a, num)
X,Y = np.meshgrid(x, y)

Z = []
Z = X**2-Y**2

for m1 in range(num):
    for m2 in range(num):
        if Y[m1][m2] >= 0:
            Z[m1][m2] = X[m1][m2]**2-Y[m1][m2]**2

fig = plt.figure()
ax = plt.axes()
plt.pcolor(X, Y, Z, cmap='ocean')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="ocean")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=20, azim=-30)

plt.show()
```
