```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-2.0, 2.5, 0.01) #xを作成
y1 = x*x #3次関数を作成

fig, ax = plt.subplots()
ax.plot(x, y1, color='black') #3次関数は黒色で曲線を描写
#fill_betweenで関数と、cos関数が大きい時黄色、3次関数が大きい時赤色で塗り潰す
ax.fill_between(x, y1, where=y1 > x*x, facecolor='blue', alpha=0.5)
plt.hlines(0, -2, 2.5, color='black')


plt.savefig("logo15.png")
plt.show()
```
