```python
## 3D2

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from mpl_toolkits.mplot3d import axes3d

plt.rcParams['figure.figsize'] = [6, 5]

num = 301
a = 5
x = np.linspace(0, 1*a ,num)
y = np.linspace(-1*a, 1*a, num)
X,Y = np.meshgrid(x, y)

Z = []
Z = 1 - np.sqrt(X)

for m1 in range(num):
    for m2 in range(num):
        if 1- np.sqrt(X[m1][m2]) <= np.abs(Y[m1][m2]):
            Z[m1][m2] = np.abs(Y[m1][m2])

fig = plt.figure()
ax = plt.axes()
plt.pcolor(X, Y, Z, cmap='ocean')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar()
#plt.savefig("Fig1-1")

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap="ocean")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(elev=-20, azim=120)
#plt.savefig("Fig1-2")

plt.show()
```
